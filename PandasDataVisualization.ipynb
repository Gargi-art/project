### Data Pre-processing
- Remove/Replace Null
- Remove Duplicate
- Change Data Type
- Remove Outlier

```python
df.isnull().sum()
```

```python
# Replacing null values in duration with mean duration
mean_duration = round(df['duration'].mean(), 2)
# mean_duration
df['duration'] = df['duration'].fillna(mean_duration)
df.isnull().sum()
```

```python
# Duplicate
df[df.duplicated()]
```

```python
# To delete duplicates
# df.drop_duplicates(inplace = True)
```

#### Change data type

```python
# Changing data type of 'price'
df['price'] = pd.to_numeric(df['price'], errors = 'coerce')
df.info()
```

```python
df.isnull().sum()
```

```python
# Dropping irrelevant column
df.drop(columns = ['Unnamed: 0', 'date'], axis = 1, inplace = True)
df.head()
```

```python
# Dropping all rows corresponding to null values
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.describe()
```

```python
# Importing plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Boxplot
fig, ax = plt.subplots(1,3, figsize = (15, 8))
sns.boxenplot(y = df['duration'], ax = ax[0])
sns.boxenplot(y = df['days_left'],ax = ax[1])
sns.boxenplot(y = df['price'], ax = ax[2])
plt.show()
```

### Outliers Removal

```python
# Remove Outlier from 'price' column
q1_price = np.quantile(df['price'], 0.25)
q3_price = np.quantile(df['price'], 0.75)
print('Quartile1 is:', q1_price)
print('Quartile3 is:', q3_price)
iqr_price = q3_price - q1_price
min_price = q1_price - 1.5 * iqr_price
max_price = q3_price + 1.5 * iqr_price
print('Min price is:', min_price)
print('Max price is:', max_price)

# Removing outlier
df = df[(df['price'] >= min_price) & (df['price'] <= max_price)]
df.shape
```
